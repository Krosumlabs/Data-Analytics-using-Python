```python
from scipy import stats

x = [6,8,9,10,12,3,18,3,5,12,13,9,7,50]
y = [99,66,87,88,120,76,95,78,67,86,66,45,66,100]

stats.linregress(x,y)
```

```python
slope,intercept,r,p,stderr = stats.linregress(x,y)
```

```python
def fx(x):
    return slope * x + intercept
```

```python
model = []
for var in x:
    r = fx(var)
    model.append(r)
    
model
```

```python
plt.scatter(x,model,c='r')
```

```python
plt.plot(x,model,c='r',marker='o')
```

```python
iris = sn.load_dataset('iris')
iris = iris[['petal_length','petal_width']]
X = iris['petal_length']
Y = iris['petal_width']
```

```python
#help(train_test_split)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.4,random_state=23)
```

```python
x_train = np.array(x_train).reshape(-1,1)
x_test = np.array(x_test).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
print(LinearRegression)
```

```python
model_obj = LinearRegression()
```

```python
model_obj.fit(x_train,y_train)
m = model_obj.intercept_
c = model_obj.coef_
```

```python
y_pred_train = m * x_train + c 
y_pred_train = model_obj.predict(x_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pred_train,color='r',marker='o')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_pred_train)
```
